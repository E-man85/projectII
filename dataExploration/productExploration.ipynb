### Column product_id

```python
# Coluna product_id duplicados
print(product['product_id'].duplicated().any())
# contar produtos únicos
print(product['product_id'].nunique())
# Ver 10 registos aleatórios
product['product_id'].sample(10)
```

### Column product_length

```python
#Valores em falta
print(product['product_depth'].isnull().any())
# Valores duplicados
print(product['product_length'].duplicated().any())
# contar produtos únicos
print(product['product_length'].nunique())
# ver estatisticas 
print(product['product_length'].describe())
# Ver 10 registos aleatórios
product['product_length'].sample(10)
```

```python
# Ver registos null
product[product['product_depth'].isnull()]
```

### Column product_width

```python
#Valores em falta
print(product['product_width'].isnull().any())
# Valores duplicados
print(product['product_width'].duplicated().any())
# contar produtos únicos
print(product['product_width'].nunique())
# ver estatisticas 
print(product['product_width'].describe())
# Ver 10 registos aleatórios
product['product_width'].sample(10)
```

### Column cluster_id

```python
#Valores em falta
print(product['cluster_id'].isnull().any())
# Valores duplicados
print(product['cluster_id'].duplicated().any())
# contar produtos únicos
print(product['cluster_id'].nunique())
# ver estatisticas 
print(product['cluster_id'].describe())
# Ver 10 registos aleatórios
product['cluster_id'].sample(10)
```

```python
# frequencias de cluster_id
product.groupby(['cluster_id']).size().sort_values(ascending=False)
```

```python
# ver NAN
product[product['cluster_id'].isnull()]
```
