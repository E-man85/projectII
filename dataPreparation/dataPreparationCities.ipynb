# Data Preparation Cities File

```python
# Carregar ficheiro
import pandas as pd
url = 'https://raw.githubusercontent.com/E-man85/projectII/main/raw_data/cities.csv'
cities = pd.read_csv(url)
cities.info()
```

## Correção dos nomes da coluna city_code  e acrescentar colunas latitude e longitude

```python
# Alteração de carecteres incorretos
# dicionario com nomes corretos
city_mapping = {
    'Denizli': 'Denizli',
    'Ankara': 'Ankara',
    '?zmir': 'İzmir',
    'Bursa': 'Bursa',
    'Antalya': 'Antalya',
    'Konya': 'Konya',
    'Adana': 'Adana',
    'Sanl?urfa': 'Şanlıurfa',
    'Gaziantep': 'Gaziantep',
    'Mersin': 'Mersin',
    'Diyarbak?r': 'Diyarbakır',
    'Kayseri': 'Kayseri',
    'Samsun': 'Samsun',
    'Istanbul': 'Istanbul',
    'Eski?ehir': 'Eskişehir',
    'Adapazar?': 'Adapazarı',
    'Kahramanmaras': 'Kahramanmaraş',
    'Erzurum': 'Erzurum',
    'Van': 'Van'
}
# Criar dataframe com lista de cities unique
city_df = pd.DataFrame(cities['city_code'].unique(), columns=['city_code'])
# Adicionar a coluna "city_code1" com os nomes corretos das cidades
city_df['city_code1'] = city_df['city_code'].map(city_mapping)
# criar aplicação para saber longitude e latitude para cada cidade
# instalar a biblioteca pip install geopy
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="app_lat_long")
# Função para obter as coordenadas de latitude e longitude
def get_coordinates(city):
    location = geolocator.geocode(city + ', Turkey')
    if location:
        return location.latitude, location.longitude
    else:
        return None, None
# Aplicar a função às cidades no dataframe
city_df['latitude'], city_df['longitude'] = zip(*city_df['city_code1'].apply(get_coordinates))
# Join com dataframe cities1 e city_df
cities =cities.merge(city_df,on='city_code',how='left')
# Eliminar coluna city_code
cities = cities.drop('city_code', axis=1)
# Renomear coluna city_code1 para city_code
cities = cities.rename(columns= {'city_code1':'city_code'})
cities.head()
```
